```python
%matplotlib inline
import numpy as np
from IPython import display
import torch
import matplotlib.pyplot as plt
```

```python
def f(x):
  return 3 * x ** 2 - 4 * x
```

```python
def num_lim(f, x, h):
  return (f(x + h) - f(x)) / h
h = 0.1
for i in range(5):
  print(f'h = {h:.5f}, num_lim:{num_lim(f, 1, h):.5f}') 
  h *= 0.1 
```

```python
display.set_matplotlib_formats('svg')
```

```python
plt.figure(figsize=(3.5, 2.5))
```

```python
x = np.arange(0, 3, 0.1)
plt.plot(x, f(x), x, 2 * x - 3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','tangent line'])
plt.show()
```
