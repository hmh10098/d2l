```python
import numpy as np
import torch
from torch.utils import data
```

```python
def synthetic_data(w, b, num_examples):
  X = torch.normal(0, 1, size = (num_examples, len(w)))
  y = torch.matmul(X, w.reshape(len(w), 1)) + b
  y += torch.normal(0, 0.01, y.shape)
  return X, y.reshape(num_examples, 1)
```

```python
true_w = torch.tensor([2, -3.4])
true_b = 4.2
X_in, y_labels = synthetic_data(true_w, true_b, 1000)
```

```python
def load_array(data_arrays, batch_size, is_Train=True):
  dataset = data.TensorDataset(*data_arrays)
  return data.DataLoader(dataset, batch_size=batch_size, shuffle=is_Train)

batch_size = 10
data_iter = load_array((X_in, y_labels), batch_size)
```

```python
iter(data_iter)
```

```python
from torch import nn
```

```python
net = nn.Sequential(nn.Linear(2, 1))
```

```python
net[0].weight.data.normal_(0, 0.01)
net[0].bias.data.fill_(0)
```

```python
loss = nn.MSELoss(reduction='sum')
trainer = torch.optim.SGD(net.parameters(), lr = 0.03)
```

```python
num_epochs = 3
for epoch in range(num_epochs):
  for X, y in data_iter:
    l = loss(net(X), y)
    trainer.zero_grad()
    l.backward()
    trainer.step()
  l = loss(net(X_in), y_labels)
  print(f'epoch: {epoch+1}, loss: {l:.5f}')
```
