```python
import torch
from torch.distributions import multinomial
```

```python
fair_prob = torch.ones([6])/6
counts = multinomial.Multinomial(10, fair_prob).sample((500,))
```

```python
c_sum = counts.cumsum(dim=0)
estimates = c_sum / c_sum.sum(dim=1, keepdim=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 4))
for i in range(6):
  plt.plot(estimates[:, i].numpy(), label= str(i))
plt.axhline(y=0.167, color='black', linestyle='dashed')
plt.xlabel('Group estimates')
plt.ylabel('estimates')
plt.legend()
```
